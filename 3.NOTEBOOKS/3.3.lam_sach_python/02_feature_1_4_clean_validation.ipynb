## 2. Raw vs Clean row count comparison

```python
df = pd.read_sql("""
    SELECT 'tracks'  AS entity,
           (SELECT COUNT(*) FROM raw.raw_tracks)  AS raw_count,
           (SELECT COUNT(*) FROM clean.tracks)    AS clean_count
    UNION ALL
    SELECT 'artists',
           (SELECT COUNT(*) FROM raw.raw_artists),
           (SELECT COUNT(*) FROM clean.artists)
""", conn)
df['match'] = df['raw_count'] == df['clean_count']
display(df)
```

## 9. release_precision distribution

```python
df_prec = pd.read_sql("""
    SELECT release_precision, COUNT(*) AS cnt,
           ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (), 2) AS pct
    FROM clean.tracks
    GROUP BY release_precision
    ORDER BY cnt DESC
""", conn)
display(df_prec)

invalid = pd.read_sql("""
    SELECT COUNT(*) AS invalid FROM clean.tracks
    WHERE release_precision NOT IN ('day','month','year','unknown') OR release_precision IS NULL
""", conn)
print('Invalid precision count:', invalid['invalid'].iloc[0])
```

## 10. Tempo NULL check

```python
df_t = pd.read_sql("""
    SELECT
        COUNT(*) FILTER (WHERE tempo IS NULL)  AS tempo_null,
        COUNT(*) FILTER (WHERE tempo <= 0)     AS tempo_lte_zero,
        COUNT(*) FILTER (WHERE tempo > 0)      AS tempo_valid,
        COUNT(*)                               AS total
    FROM clean.tracks
""", conn)
display(df_t)
```
